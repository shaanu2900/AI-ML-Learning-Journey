```python
#Basics
print("Hello World")

#Variables
name = "Shaantanu"
surname = "Kadam"
age = 25
print(name, surname, "is of age", age, ".")
print(name, "is a Genius.")
is_adut = True

#Concatenation
study = input("What do you study? - ")
print("Thats great to know that you are studying "+study)
power = input("What is your secret super power? - ")
print("So, ",name," is a Secret Superhero and has the superpower of ",power)

#Type Conversion
first = int(input(" Enter first number : "))
second = int(input(" Enter second number : "))
sum = int(first) + int(second)
print("The sum of {0} and {1} is {2}" .format(first, second, sum))

#Strings
print(name.upper())

print(name.find('u'))
print(name.replace("Shaantanu","Shaanu"))
print('S' in name)

#Arithmetic Operstors
print(5+2)
print(5-2)
print(5*2)
print(5/2)
print(5//2)
print(5**2)

#Shortcuts
i = 5
i = i + 2
i += 2
i -= 2
i *= 2

#Comparison Operators
print(3>2)
print(3>=2)
print(3<2)
print(3<=2)
print(3==2)
print(3!=2)

#Logical Operators
print(2>3 or 2>1)
print(2>3 and 2>1)
print(not 2>3 and 2>1)

#If else
if age >= 18:
  print("You are an adult")
elif age >= 13:
  print("You are a teenager")
else:
  print("You are a chid")

#Range
numbers = range(5)
print(numbers)

#While Loop
j = 1
while j <= 5:
  print(j * "*")
  j += 1

#For Loop
for item in range(5):
  print(item)
  print(item + 1)

#List
marks = [95,89,91,32,57]
print(marks)
print(marks[0])
print(marks[-1])
print(marks[1:3])

#Append
marks.append(99)
print(marks)

#Insert
marks.insert(0,99)
print(marks)

#Extend
marks.extend([100,100,100])
print(marks)

#Length
print(len(marks))

#Count
print(marks.count(100))

#Sort
marks.sort()
print(marks)

#Reverse
marks.reverse()
print(marks)

#Pop
marks.pop()
print(marks)

#Clear
marks.clear()
print(marks)

#Break & Continue
students = ["Shaantanu","Yathi","Raj","Siddharth","Atharva","Yamin"]
for student in students:
  if student == "Raj":
    continue
  elif student == "Atharva":
    break
  print(student)

#Tuples
marks = (95,89,91,32,57,95,89,95)
print(marks)
print(marks[0])
print(marks[-1])
print(marks[1:3])
print(marks.count(95))
print(marks.index(91))

#Sets
marks = {95,89,91,32,57,95,89,95}
print(marks)
marks.add(99)
print(marks)
marks.remove(99)
print(marks)

#Dictionary
marks = {"Shaantanu":95,"Yathi":89,"Raj":91,"Siddharth":32,"Atharva":57}
print(marks)
print(marks["Shaantanu"])
marks["Yamin"] = 47
print(marks)
marks["Shaantanu"] = 99

#Functions
def greet(name):
  print("Hi",name)
greet("Shaantanu")

def print_sum(first,second):
  print(first + second)
print_sum(5,2)
```
